### Details du projet : Speed Dating

**But**:  Votre but va être de comprendre ce qui se passe lors d'un speed dating et surtout de comprendre ce qui va influencer l'obtention d'un deuxième rendez-vous.

Il faut comprendre que le dataset à été crée a partir de vagues successives pour etablir des groupes (par exemple, la 1ere vague comprenait un groupe de 20 personnes : 10 hommes et 10 femmes)

**À propos des données** :

Les données ont été recueillies auprès de **552 participants** à des événements expérimentaux de speed dating de 2002 à 2004.

Il existe 21 cessions(waves) qui consiste à réunir un même nombre de participants des 2 genres(hommes et femmes)

Pendant ces cessions, les participants ont un "premier rendez-vous" de 4 minutes avec tous les autres participants du sexe opposé.

À la fin de leurs quatre minutes, on a demandé aux participants s'ils aimeraient revoir leur partenaire et ont également été invités à les évaluer sur six critères :

- **`Attraction`**
- **`Sincérité`**
- **`Intelligence`**
- **`Fun`**
- **`Ambition`**
- **`Centres d'interets`**

L'ensemble de données comprend également des données de questionnaire recueillies auprès des participants à différents stades du processus. Ces champs incluent :

- Etude démographie sur certaines caractèristiques (origine ethnique, localisation, evolution professionnelle,ect...)
- Habitudes de rencontres
- Perception de soi à travers des attributs clés
- Interets des valeurs qui sont importantes chez un partenaire
- Informations sur le mode de vie

```python
def proportion_nan_values(df, num=None):
    n = len(df)
    nan_vals = df.columns[df.isna().sum() == n]
    if not num:      
        print(f"Colonne(s) entièrement remplie(s) de valeurs manquantes :\n\n {nan_vals}")
        return nan_vals
    almost_nan = df.columns[(df.isna().sum() / n > num) & (df.isna().sum() != n)]
    print(f"Colonne(s) remplie(s) au moins de {num*100}% de valeurs manquantes (sans les NaN à 100%):\n\n {almost_nan}")
    return almost_nan
```
