## C et Gamma pour Support Vector Machine

### Objectif SVM

Il permet de trouver le meilleur point (en 1-D), la meilleure ligne (en 2-D), le meilleur plan (3-D), ou le meilleur hyperplan (en plus de 3-D) pour séparer les classes : ce qui permet de séparer ces classes s'apelle la **Frontière de Decision** (**Decision Boundary**)

Lorsqu'il s'agit d'un ensemble séparable, cad que l'on peut linéairement séparer 2 classes, cela va suffir. Mais lorsque ces deux classes ne sont pas linéairement séparables, nous allons devoir trouver une technique pour y arriver : le **kernel trick**

**Support Vectors** : ce sont les points les plus proches de l'hyperplan. Une ligne de séparation sera définie à l'aide de ces points de données.

**Margin** : c'est la distance entre l'hyperplan et les observations les plus proches de l'hyperplan (vecteurs supports). Dans SVM, une grande marge est considérée comme une bonne marge. Il existe deux types de marges, la hard margin et la soft margin. 

Le but de SVM est de minimiser la marge d'erreur de classification des deux classes : en choisissant la distance entre l'hyperplan de séparation et les vecteurs de supports (cette distance est la **margin**), nous allons pouvoir minimiser ces erreurs.

### Hyperparamètre `C`

`C` est un hypermètre dans SVM pour **contrôler l'erreur de la margin**.

Plus `C` est grand, plus la marge d'erreur est grande, plus `C` est petit, plus la marge d'erreur est petite.

Il faut faire attention car si nous choisissons un `C` trop petit, il y aura moins d'erreur mais sa capacité à généraliser le modèle sur le test va se retrouver abaisser car en sur-adaptant la frontière de décision pour minimiser les erreurs, SVM risque de créer des frontières trop complexe qui vont créer de l'overfitting sur le train et des baisse de performance sur le test.

Donc, avoir de faibles erreurs ne signifie pas d'avoir un meilleur modèle !

### Hyperparamètre `gamma`

**Gamma** sert pour **donner le poids de courbure de la limite de décision** est utilisé lorsque nous utilisons le **noyau RBF gaussien**. si vous utilisez un noyau linéaire ou polynomial, vous n'avez pas besoin de gamma, vous avez seulement besoin d'un hypermètre C. Quelque part, il est également utilisé comme sigma. En fait, sigma et gamma sont liés par cette relation:

$$\text{GAUSSIAN RBF} = \exp \frac{-1 ||X - X`||}{2\sigma^2}$$

$$(Gamma)\space Y = \frac {1}{2\sigma^2}$$

$$\text{GAUSSIAN RBF} = \exp (-Y ||X - X`||)$$

**Gamma** = **sigma**.

Gamma est un hyperparamètre que nous devons définir avant d'entraîner le modèle. Gamma décide de la courbure que nous voulons dans une limite de décision :

* **Gamma élevé** signifie **plus de courbure**.
* **Gamma bas** signifie **moins de courbure**.

Pour savoir comment régler `C` et `gamma`, on va faire une gridsearch avec ces paramètres :

`C`: [0.001, 0.01, 0.1, 1, 10, 100]</br>
`Gamma` :  [0.001, 0.01, 0.1, 1, 10, 100]

### `Support Vector Machine` VS `Logistic Regression` pour séparer deux classes

Selon le nombre de features dont on dispose, on peut choisir la régression logistique ou SVM.

**SVM fonctionne mieux** lorsque l'ensemble de données est **petit et complexe**. Il est généralement conseillé d'utiliser d'abord la régression logistique et de voir comment elle fonctionne, si elle ne donne pas une bonne précision, vous pouvez opter pour SVM sans aucun kernel. La régression logistique et la SVM sans kernel ont des performances similaires mais selon vos features, l'un peut être plus efficace que l'autre.
